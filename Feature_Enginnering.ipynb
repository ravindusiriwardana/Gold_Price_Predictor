```python
import pandas as pd

# 1. Define the input path
input_path = r"C:\\Users\\Admin\\Desktop\\Gold_Price_Predictor\\gold_features_clean_2015_2026_02.xlsx"

# Load the data
print("Loading data...")
df = pd.read_excel(input_path)

# 2. Format and sort by date (Crucial for time-series)
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values(by='Date').reset_index(drop=True)

# 3. Dynamically identify features to lag
cols_to_exclude = ['Date', 'Month', 'Quarter']
features_to_lag = [col for col in df.columns if col not in cols_to_exclude and 'Target' not in col]

print(f"Applying lags to {len(features_to_lag)} variables...")

# 4. Define the lag intervals
lags = [10, 30, 60, 90]

# 5. Generate the lagged variables programmatically
for col in features_to_lag:
    for lag in lags:
        df[f'{col}_lag_{lag}'] = df[col].shift(lag)

# 6. Create the 90-day ahead target variable
if 'Gold_Close' in df.columns:
    df['Target_Price_90D'] = df['Gold_Close'].shift(-90)
else:
    print("Warning: 'Gold_Close' column not found. Target variable not created.")

# 7. Clean up the dataset
print("Cleaning missing values caused by lagging...")
model_df = df.dropna().reset_index(drop=True)

# 8. Define the output path and save the new dataset
output_path = r"C:\Users\Admin\Desktop\Gold_Price_Predictor\gold_features_lagged_90D.xlsx"

print("Saving new dataset...")
model_df.to_excel(output_path, index=False)

# Print final stats
print("-" * 30)
print(f"Success! File saved to:\n{output_path}")
print("-" * 30)
print(f"Original Data Shape: {df.shape}")
print(f"Final Model Data Shape: {model_df.shape}")
```

```python
"""
╔══════════════════════════════════════════════════════════════════════╗
║       GOLD CLOSE — FUTURE-PRICE LAG GENERATOR (10 / 30 / 60 / 90d)   ║
╚══════════════════════════════════════════════════════════════════════╝

Purpose
-------
Loads the daily gold feature dataset already produced by the scraper
(gold_features_clean_2015_2026_daily.xlsx) and adds columns showing
the gold price N DAYS IN THE FUTURE, attached to each row's date.

Example
-------
  Row for 2015-01-01  ->  Gold_Close_Lag_10 = Gold_Close value on 2015-01-11
  Row for 2015-01-01  ->  Gold_Close_Lag_30 = Gold_Close value on 2015-01-31
  Row for 2015-01-01  ->  Gold_Close_Lag_60 = Gold_Close value on 2015-03-02
  Row for 2015-01-01  ->  Gold_Close_Lag_90 = Gold_Close value on 2015-03-31

This is the format you want when every OTHER column (DXY, VIX, CPI, etc.
as of 2015-01-01) is being used to predict what gold will be worth N
days later — each row already lines up "today's inputs" next to
"gold price N days from today" in the same row, ready for modeling.

What it does
------------
  - Gold_Close_Lag_10  = Gold_Close, 10 days AFTER the row's date
  - Gold_Close_Lag_30  = Gold_Close, 30 days AFTER the row's date
  - Gold_Close_Lag_60  = Gold_Close, 60 days AFTER the row's date
  - Gold_Close_Lag_90  = Gold_Close, 90 days AFTER the row's date

All other columns are left completely untouched.
NO rows or dates are dropped — the LAST 10/30/60/90 rows will have
NaN in the corresponding lag column, because there's no future price
data available that far past the end of the dataset yet.
"""

import pandas as pd

# =====================================================================
# CONFIGURATION
# =====================================================================
INPUT_FILE   = "C:\\Users\\Admin\\Desktop\\Gold_Price_Predictor\\gold_features_clean_2015_2026_02.xlsx"
OUTPUT_FILE  = "gold_features_clean_2015_2026_daily_LAGGED.xlsx"
LAG_PERIODS  = [10, 30, 60, 90]     # days ahead to pull Gold_Close from
PRICE_COLUMN = "Gold_Close"
# =====================================================================

def add_future_price_columns(df: pd.DataFrame, price_col: str, lags: list) -> pd.DataFrame:
    """
    For each value in `lags`, adds a column holding `price_col`'s value
    N days AFTER the current row's date (a "lead", pulled back onto the
    current row). Does NOT drop or reindex any rows — NaNs are left in
    place for the trailing rows where a given future value doesn't exist
    yet (i.e. within N days of the end of the dataset).
    """
    df = df.copy()

    if price_col not in df.columns:
        raise KeyError(f"Column '{price_col}' not found in the dataset.")

    for lag in lags:
        col_name = f"{price_col}_Lag_{lag}"
        # shift(-lag) pulls the value from `lag` rows AHEAD back onto this row
        df[col_name] = df[price_col].shift(-lag)
        print(f"    OK  {col_name:25s} -> {price_col}, {lag} day(s) ahead of this row's date")

    return df


def main():
    print(f"\n[1/3] Loading dataset: {INPUT_FILE} ...")
    df = pd.read_excel(INPUT_FILE)

    # Ensure Date is a proper datetime and sorted chronologically
    # (lag features only make sense on a correctly time-ordered series)
    if "Date" in df.columns:
        df["Date"] = pd.to_datetime(df["Date"])
        df = df.sort_values("Date").reset_index(drop=True)

    print(f"    Rows loaded  : {len(df):,}")
    print(f"    Date range   : {df['Date'].min().date()}  to  {df['Date'].max().date()}")

    print(f"\n[2/3] Creating future-price columns for '{PRICE_COLUMN}' ({LAG_PERIODS} days ahead) ...")
    df_lagged = add_future_price_columns(df, PRICE_COLUMN, LAG_PERIODS)

    print(f"\n[3/3] Saving output -> {OUTPUT_FILE} ...")
    df_lagged.to_excel(OUTPUT_FILE, index=False)

    print("\n" + "=" * 60)
    print("DONE")
    print("=" * 60)
    print(f"  Original columns : {len(df.columns)}")
    print(f"  New columns      : {len(df_lagged.columns)}  (+{len(LAG_PERIODS)} lag columns)")
    print(f"  Rows (unchanged) : {len(df_lagged):,}")
    for lag in LAG_PERIODS:
        col = f"{PRICE_COLUMN}_Lag_{lag}"
        missing = df_lagged[col].isna().sum()
        print(f"    {col:25s} -> {missing} trailing NaN row(s) (expected, no future data that far past dataset end)")
    print(f"\n  File saved to: {OUTPUT_FILE}")


if __name__ == "__main__":
    main()
```
